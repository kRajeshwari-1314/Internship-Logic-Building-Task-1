```python
#Problem 1: Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    # Checking post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status
# Sample Data
likes = [200, 150, 300, 400]
total, post_status = analyze_engagement(likes)
print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    return status
# Sample Data
medicine_stock = 6
print("Medicine Stock:", medicine_stock)
print("Status:", check_medicine_stock(medicine_stock))
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):
    total = 0
    # Calculating total rainfall
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    # Checking adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average, status
# Sample Data
rainfall = [70, 80, 75, 65, 70]
required = 70
avg, rain_status = rainfall_checker(rainfall, required)
print("Average Rainfall:", avg)
print("Rainfall Status:", rain_status)
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def detect_duplicates(usernames):
    # Converting list to set to remove duplicates
    unique_users = set(usernames)
    if len(unique_users) != len(usernames):
        return "Yes"
    else:
        return "No"
# Sample Data
users = ["snehal01", "rahul22", "snehal01", "amit45"]
print("Duplicate Accounts Found:", detect_duplicates(users))
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker

def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return status
# Sample Data
patient_age = 21
print("Patient Age:", patient_age)
print("Eligibility Status:", check_eligibility(patient_age))
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter

def premium_crop_filter(price_list):
    premium_crops = []
    # Filtering prices above 2000
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    return premium_crops
# Sample Data
crop_prices = [1500, 2500, 1800, 3200, 2100]
print("Premium Crops:", premium_crop_filter(crop_prices))
```

```python
# Problem 7: System Monitoring – Application Health Checker

def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return status
# Sample Data
error_count = 7
print("Error Count:", error_count)
print("System Status:", application_health(error_count))
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker

def transaction_checker(amount):
    daily_limit = 50000
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    return status
# Sample Data
transaction_amount = 60000
print("Transaction Amount:", transaction_amount)
print("Transaction Status:", transaction_checker(transaction_amount))
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    # Counting attended classes
    for day in attendance_list:
        if day == 1:
            attended += 1
    percentage = (attended / total_classes) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
# Sample Data (1 = Present, 0 = Absent)
attendance_data = [1, 1, 1, 0, 1, 1, 1, 1]
percent, eligibility = attendance_checker(attendance_data)
print("Attendance Percentage:", percent)
print("Exam Eligibility:", eligibility)
```

```python
#Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
    bill = 0
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status
# Sample Data
units_consumed = 250
total_bill, usage = electricity_bill(units_consumed)
print("Total Bill:", total_bill)
print("Usage Classification:", usage)
```
